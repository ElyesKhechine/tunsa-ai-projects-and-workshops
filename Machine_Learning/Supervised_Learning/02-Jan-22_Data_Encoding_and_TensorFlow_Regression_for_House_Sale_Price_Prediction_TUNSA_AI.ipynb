```python
linear_model = tf.keras.Sequential([
    normalizer,
    tf.keras.layers.Dense(units=1)
])
```

```python
linear_model.compile(
    optimizer=tf.optimizers.Adam(learning_rate=0.1),
    loss='mean_absolute_error')
```

```python
history = linear_model.fit(
    train,
    label,
    epochs=100,
    verbose=2,
    validation_split = 0.2)
```
