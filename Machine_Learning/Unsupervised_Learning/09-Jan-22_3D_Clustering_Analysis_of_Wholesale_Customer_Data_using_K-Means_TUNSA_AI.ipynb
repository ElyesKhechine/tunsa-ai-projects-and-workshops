```python
import numpy as np
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import pandas as pd
data=pd.read_csv('/content/Wholesale customers data.csv')
```

```python
data.head()
```

```python
data.info()
```

```python

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.array(data['Fresh'])
y = np.array(data['Milk'])
z = np.array(data['Grocery'])

ax.scatter(x,y,z, marker="s", c='r', s=40, cmap="RdBu")
```

```python
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
fig = plt.figure(figsize = (30, 30))
ax = plt.axes(projection ="3d")
ax.scatter(xs=data['Fresh'], ys=data['Milk'],zs=data['Grocery'], c='r')
plt.title("simple 3D scatter plot")
```

```python
fig, ax = plt.subplots()

ax.scatter(x=data['Fresh'], y=data['Milk'])
```

```python
model = KMeans(n_clusters=2, init='random', random_state=42, max_iter=100, n_init=1)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data['Fresh'], y=data['Milk'], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```

```python
model = KMeans(n_clusters=8, init='k-means++', random_state=42, max_iter=100, n_init=10)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data['Frozen'], y=data['Detergents_Paper'], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```
