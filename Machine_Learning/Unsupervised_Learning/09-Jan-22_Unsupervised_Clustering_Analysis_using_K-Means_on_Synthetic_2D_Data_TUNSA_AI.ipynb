```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Generate some numbers
data = np.random.randn(600,2)
data[:100] += 2,3
data[100:200] += 5, 3
data[200:300] += 3, 5
data[300:] += 2, 7
```

```python
fig, ax = plt.subplots()

ax.scatter(x=data[:,0], y=data[:,1])
```

```python
model = KMeans(n_clusters=4, init='random', random_state=42, max_iter=100, n_init=1)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data[:,0], y=data[:,1], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```

```python
model = KMeans(n_clusters=8, init='k-means++', random_state=42, max_iter=100, n_init=10)

model.fit(data)

y_pred = model.predict(data)

fig, ax = plt.subplots()
ax.scatter(x=data[:,0], y=data[:,1], c=y_pred)
ax.scatter(x=model.cluster_centers_[:,0], y=model.cluster_centers_[:,1], c='r')
```
