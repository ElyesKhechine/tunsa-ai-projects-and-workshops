```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def elem(column): 
  l=[]
  c=list(set(column))
  for i in range(len(c)) :
    l.append(list(column).count(c[i]))
  return(max(l))
```

```python
nb={}
for i in df.columns :
  nb[i]=(elem(df[i])/np.size(df,axis=0))*100 
  if nb[i]>85:
    nb.pop(i)
    df=df.drop(i,axis=1)
```
